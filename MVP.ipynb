# Predicting IMDB Rating on Movies

The goal for this project is to predict IMDB ratings by using Linear Regression. 

Here is the target value and the categorial/numerical features. 
   - Target : IMDB Ratings
   - Features : Genre, Content Rating, Director, Stars, Release Year, Runtime, Votes, Metascore, Domestic Gross Earning, IMDB Rating
